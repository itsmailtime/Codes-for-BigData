```python
names=[]
starStr=[]
numReviewsStr=[]
cuisine=[]
priceRangeStr=[]


for restaurant in driver.find_elements(By.CLASS_NAME, "YHnoF.Gi.o"):
    if(not (bool(re.search("[0-9]+", restaurant.get_attribute("data-test"))))):
        continue
    names.append(restaurant.find_element(By.CLASS_NAME, "Lwqic.Cj.b").text)
    starStr.append(restaurant.find_element(By.CLASS_NAME, "UctUV.d.H0").get_attribute("aria-label"))
    numReviewsStr.append(restaurant.find_element(By.CLASS_NAME, "IiChw").text)
    cuisine.append(None)
    priceRangeStr.append(None)
    for extraInfo in restaurant.find_element(By.CLASS_NAME, "nrKLE.PQvPS.bAdrM").find_elements(By.CLASS_NAME, "ABgbd"):
        if("$" in extraInfo.text):
            priceRangeStr[-1]=extraInfo.text
        else:
            cuisine[-1]=extraInfo.text
    
print(names)
print(starStr)
print(numReviewsStr)
print(cuisine)
print(priceRangeStr)

print(len(names))
print(len(starStr))
print(len(numReviewsStr))
print(len(cuisine))
print(len(priceRangeStr))

df = pd.DataFrame({"Name":names, "Cuisine":cuisine, "Review":numReviewsStr, "Point":starStr, "Price":priceRangeStr})
df
```

```python
def getReviews(r=10): # range is *30, TODO: make it so you can do any city
    names=[]
    star=[]
    numReviews=[]
    cuisine=[]
    priceRange=[]
    for i in tqdm(range(0, r)):
        driver.get(f"https://www.tripadvisor.com/Restaurants-g32655-oa{i*30}-Los_Angeles_California.html#EATERY_LIST_CONTENTS")
        for restaurant in driver.find_elements(By.CLASS_NAME, "YHnoF.Gi.o"):
            if(not (bool(re.search("[0-9]+", restaurant.get_attribute("data-test"))))):
                continue
            names.append(restaurant.find_element(By.CLASS_NAME, "Lwqic.Cj.b").text)
            star.append(float(restaurant.find_element(By.CLASS_NAME, "UctUV.d.H0").get_attribute("aria-label").split(" ")[0]))
            numReviews.append(int(restaurant.find_element(By.CLASS_NAME, "IiChw").text.split(" ")[0].replace(",", "")))
            cuisine.append(None)
            priceRange.append(None)
            for extraInfo in restaurant.find_element(By.CLASS_NAME, "nrKLE.PQvPS.bAdrM").find_elements(By.CLASS_NAME, "ABgbd"):
                if("$" in extraInfo.text):
                    extraInfo.text.split
                    if(len(extraInfo.text.split(" - "))==1):
                        priceRange[-1]=2*[len(extraInfo.text)]
                    elif(len(extraInfo.text.split(" - "))==2):
                        priceRange[-1]=[len(price) for price in extraInfo.text.split(" - ")]
                else:
                    cuisine[-1]=extraInfo.text
    return pd.DataFrame({"name":names, "stars":star, "numReviews":numReviews, "cuisine":cuisine, "priceRange":priceRange})
```

```python
df = getReviews(r=1000)
```
