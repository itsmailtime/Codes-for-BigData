```python
def wordCount(df,num):
    total_word = []
    for i in df['review'].apply(lambda x: re.findall('[a-z]+',x.lower())):
        total_word.extend(i)
    return Counter(total_word).most_common(num)
```

```python
result = wordCount(df,30)
stopwords = [result[i][0] for i in range(30)]
stopwords
```

```python
def wordCount_stopwrods(df,num):
    total_word = []
    for i in df['review'].apply(lambda x: re.findall('[a-z]+',x.lower())):
        for word in i:
            if word not in stopwords:
                total_word.append(word)
    return Counter(total_word).most_common(num)
```

```python
wordCount_stopwrods(df_neg,30)
```

```python
wordCount_stopwrods(df_pos,100)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
wordcloud = WordCloud(
    background_color='black',
    colormap = 'Accent',   
    width = 1800,
    height = 800
)

wordcloud_words = wordcloud.generate_from_frequencies(dict(wordCount_stopwrods(df_pos,100)))
array = wordcloud_words.to_array()

fig = plt.figure(figsize=(100, 10))
plt.imshow(array, interpolation="bilinear")
plt.axis('off')
plt.show()
```
